```python
import whisper
import gradio as gr
import tempfile
import os
```

```python
model = whisper.load_model("base")
```

```python
def transcribe_audio(audio):

    if audio is None:
        return "No audio provided.", "", "", None

    result = model.transcribe(audio)

    transcript = result["text"]

    # Word count
    word_count = len(transcript.split())

    # Estimated speaking duration
    duration = result["segments"][-1]["end"] if result["segments"] else 0

    # Save transcript to file for download
    with tempfile.NamedTemporaryFile(delete=False, suffix=".txt") as f:
        f.write(transcript.encode("utf-8"))
        transcript_file = f.name

    return transcript, word_count, round(duration,2), transcript_file
```

```python
with gr.Blocks() as app:

    gr.Markdown("# 🎤 Whisper Speech-to-Text App")
    gr.Markdown("Upload **wav or mp3 audio** to generate transcript")

    audio_input = gr.Audio(type="filepath", label="Upload Audio (wav/mp3)")

    transcript_output = gr.Textbox(label="Transcript")

    word_output = gr.Number(label="Word Count")

    duration_output = gr.Number(label="Estimated Speaking Duration (seconds)")

    download_file = gr.File(label="Download Transcript")

    with gr.Row():
        transcribe_btn = gr.Button("Transcribe")
        clear_btn = gr.Button("Clear")

    transcribe_btn.click(
        transcribe_audio,
        inputs=audio_input,
        outputs=[transcript_output, word_output, duration_output, download_file]
    )

    clear_btn.click(
        lambda: ("", "", "", None),
        None,
        [transcript_output, word_output, duration_output, download_file]
    )

app.launch(share=True)
```

```python
import numpy as np
import pandas as pd
import gradio as gr
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = {
    "Movie1":[5,1,1,4],
    "Movie2":[1,5,3,0],
    "Movie3":[4,1,2,5],
    "Movie4":[0,4,5,0],
    "Movie5":[3,2,4,1],
    "Movie6":[5,0,1,5]
}

features = ["Action","Romance","Comedy","SciFi"]

df = pd.DataFrame(movies,index=features).T

df
```

```python
def recommend_movies(action, romance, comedy, scifi):

    user_profile = np.array([[action, romance, comedy, scifi]])

    similarity = cosine_similarity(df.values, user_profile)

    df_copy = df.copy()
    df_copy["Similarity"] = similarity

    top_movies = df_copy.sort_values(by="Similarity", ascending=False).head(3)

    return top_movies[["Similarity"]]
```

```python
with gr.Blocks() as app:

    gr.Markdown("# 🎬 AI Movie Recommender")
    gr.Markdown("Adjust your movie preferences and get recommendations")

    action = gr.Slider(0,5,label="Action Preference")
    romance = gr.Slider(0,5,label="Romance Preference")
    comedy = gr.Slider(0,5,label="Comedy Preference")
    scifi = gr.Slider(0,5,label="SciFi Preference")

    output = gr.Dataframe()

    recommend_btn = gr.Button("Recommend Movies")

    recommend_btn.click(
        recommend_movies,
        inputs=[action, romance, comedy, scifi],
        outputs=output
    )

app.launch(share=True)
```
